```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
Assignment2 =pd.read_csv('/content/Assignment2.csv')
duplicate_rows=Assignment2[Assignment2.duplicated()]
Assignment2.head(5)
print("Duplicate Rows:")
print(duplicate_rows)

#Creating Data Frame
data = pd.DataFrame({
    'accident_severity': ['Fatal', 'Major', 'Minor'],
    'speed_of_vehicle': [80, 50, 60],
    'weather_condition': ['Snowy', 'Rainy', 'Clear'],
    'road_condition': ['Slippery', 'Wet', 'Dry'],
    'driver_age': [20, 33, 41],
    'vehicle_condition': ['Fair', 'Good', 'Excellent'],
    'number_of_passengers': [8, 3, 1]
})
# Encode 'accident_severity' using map
data['accident_severity_encoded'] = data['accident_severity'].map({'minor': 3, 'major': 2, 'fatal': 1})

# Encode 'weather_condition' using map
data['weather_condition_encoded'] = data['weather_condition'].map({'Snowy': 1, 'Rainy': 2, 'Clear': 3})

# Encode 'road_condition' using map
data['road_condition_encoded'] = data['road_condition'].map({'Slippery': 1, 'Wet': 2, 'Dry': 3})

# Encode 'vehicle_condition' using map
data['vehicle_condition_encoded'] = data['vehicle_condition'].map({'Fair': 1, 'Good': 2, 'Excellent': 3})

# Now we drop the original string columns
data_numeric = data.drop(columns=['accident_severity', 'weather_condition', 'road_condition', 'vehicle_condition'])

# Display the encoded DataFrame
print(data_numeric)
data.head()
print(data)

# Checking for missing values
print(data.isnull().sum())
# Statistics for numerical columns
print(data.describe())
# Histograms
data.hist(figsize=(5, 5), bins=10)
plt.tight_layout()
plt.show()
# Correlation matrix
correlation_matrix = data_numeric.corr()

# Plot heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Boxplot for speed_of_vehicle grouped by accident_severity
plt.figure(figsize=(8, 6))
sns.boxplot(x='accident_severity', y='speed_of_vehicle', data=data)
plt.title('Vehicle Speed vs Accident Severity')
plt.show()

# Count plot for weather condition and accident severity
plt.figure(figsize=(8, 6))
sns.countplot(x='weather_condition', data=data)
plt.title('Weather Condition Distribution')
plt.show()

# Pairplot to visualize relationships
sns.pairplot(data)
plt.show()
```
